## Data Cleaning

We notice that we have 3 columns, ```released_year```, ```released_month``` and ```released_day```, which we can and have to merge together and parse this as a date.This will help us immensely, when diving deeper into the analysis part.

````python
df.dropna()
df
````

````python
df = df.drop_duplicates()
df.shape
````

````python
df = df.dropna(subset=['released_year', 'released_month', 'released_day'])
df.shape
````

````python
print(df[(df['released_year'] < 1900) | (df['released_year'] > 2024)])
print(df[(df['released_month'] < 1) | (df['released_month'] > 12)])
print(df[(df['released_day'] < 1) | (df['released_day'] > 31)])

df.info()
````

Small problems seem to appear when trying to create a ```release_date``` column, lets fix it!

````python
try:
    df['release_date'] = df.apply(lambda row: pd.to_datetime({
        'year': row['released_year'],
        'month': row['released_month'],
        'day': row['released_day']
    }), axis=1)
except Exception as e:
    print("Error in converting to datetime:", e)
````

````python
df['release_date_str'] = df['released_year'].astype(str) + '-' + df['released_month'].astype(str).str.zfill(2) + '-' + df['released_day'].astype(str).str.zfill(2)

# Convert the combined string column to datetime
try:
    df['release_date'] = pd.to_datetime(df['release_date_str'])
    df.drop(columns=['release_date_str'], inplace=True)
except Exception as e:
    print("Error in converting to datetime:", e)
    
df
````

````python
df.drop(columns = ['released_year', 'released_month', 'released_day'], inplace = True)
df
````

````python
pd.set_option('display.max_columns', None)
df
````

````python
df['in_deezer_playlists'] = pd.to_numeric(df['in_deezer_playlists'], errors='coerce')
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'], errors='coerce')
````

### Weekday Analysis Findings

We can see, that a vast majority of the songs gets released either Friday or Thursday.This has several reasons:
***
  
1. ‚ú≥Ô∏è**New Music Friday Playlists**: Spotify and other streaming services feature curated playlists like "New Music Friday," which showcase new releases. By releasing music on Thursday night (which aligns with Friday in many time zones), artists can ensure their songs are included in these highly trafficked playlists right from the start of the release period, maximizing visibility.
***
2. ‚ú≥Ô∏è**Chart Tracking**: The music industry tracks sales and streams on a weekly basis, typically from Friday to Thursday. Releasing music at the beginning of this period (Friday) gives songs the maximum amount of time to accumulate streams and sales, which can impact chart positions on platforms like the Billboard Hot 100.
***
3. ‚ú≥Ô∏è**Consumer Behavior**: People tend to have more free time to discover and listen to new music over the weekend. Releasing songs on Friday taps into this behavior, increasing the likelihood that listeners will engage with new releases when they have the time and inclination to do so.
***
4. ‚ú≥Ô∏è**Global Realease Strategy**: A simultaneous global release ensures that fans around the world can access new music at the same time, reducing the risk of piracy and ensuring a more coordinated marketing push. Releasing music on Thursday night aligns with Friday in many major markets due to time zone differences.
***
5. ‚ú≥Ô∏è**Weekend Promotions and Events**: Artists often plan promotional events, live performances, or media appearances around the time of a new release. Launching a song or album at the end of the week provides a natural lead-in to these promotional activities.

````python
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings("ignore")

# Features for clustering
features_for_clustering = ['streams', 'in_spotify_playlists', 'in_spotify_charts',
                           'in_apple_playlists', 'in_apple_charts', 'in_deezer_playlists',
                           'in_deezer_charts', 'in_shazam_charts', 'danceability_%',
                           'valence_%', 'energy_%']

# Scatter plots for relationships between features
relationships = [
    ('streams', 'in_spotify_playlists'),
    ('streams', 'in_spotify_charts'),
    ('streams', 'in_apple_playlists'),
    ('streams', 'in_apple_charts'),
    ('streams', 'in_deezer_playlists'),
    ('streams', 'in_deezer_charts'),
    ('streams', 'in_shazam_charts'),
    ('streams', 'danceability_%'),
    ('streams', 'valence_%'),
    ('streams', 'energy_%')
]

# Clustering
kmeans = KMeans(n_clusters=3)  # Nr of clusters adjustable
df['cluster'] = kmeans.fit_predict(df[features_for_clustering])

# Plotting
plt.figure(figsize=(15, 20))
for i, (x, y) in enumerate(relationships, 1):
    plt.subplot(5, 2, i)
    sns.scatterplot(data=df, x=x, y=y, hue='cluster', palette='Set2')
    plt.title(f'{x} vs {y}')
plt.tight_layout()
plt.show()
````
